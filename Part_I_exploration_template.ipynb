### showing "member_gender" and "user_type" in the relationship between "duration_sec" and "age"

for this plot i want to minimize the range of duration_sec and age to have a better representation in the scatter plot. so i created df1 which max duration_sec and age are respectively 2000 and 80


 

```python
df1 = df[(df["duration_sec"] < 2000) & (df["age"] <80)]
```

Also i took 1000 sample for each member_gender from 'df' and i concatinated the three dataset into one called "gender".

The purpose of my work is to do a fairly comparison and have a better insight.


```python
Other = df1[df1["member_gender"] == "Other"].sample(1000)
Male = df1[df1["member_gender"] == "Male"].sample(1000)
Female = df1[df1["member_gender"] == "Female"].sample(1000)

gender = pd.concat([Other, Male, Female])
```

```python
#to reduce repetitive code i created a function that create a facegrid and plot a scatter using seaborn "regplot" method.
def faceting(data, col, hue,  x, y, alpha, size  ):
    # 7 parameters are set to the faceting function :
    plot = sb.FacetGrid(data = data, col = col,hue = hue, size = size )
    #FacetGrid is an object class with specified parameters : 
    #data : data selected
    #col, hue : Variables that define subsets of the data, which will be drawn on separate facets in the grid.
    #size: size of the facet
    plot.map(sb.regplot , x, y,  x_jitter = 1, scatter_kws={'alpha':alpha}, fit_reg = False );
    # plot.map specifies what kind of plot we want to have
    #x,y variables to plot in each axis
    #alpha : set the level of transparency for the data
    plot.add_legend()
```

this function will be used below to make scatter plots

```python
faceting(data = gender, hue = "member_gender", x = "age", y = "duration_sec",col = None, alpha = 0.5, size = 6)
```

the graph is very messy and not understandable , let's plot member_gender into three graphs

```python
faceting(data = gender, col = "member_gender", x = "age", y = "duration_sec", alpha = 0.3, size = 6, hue = None  )
```

after spliting , we can see that there isn't a strong relationship between age and duration_sec for each of the member_gender,

the differences are not significant between the three scatter plot.

the three graph have high amount of data in age between [20 40], and duration_sec between [200;1000]
